```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as ss

plt.style.use("ggplot")
plt.rcParams['figure.figsize'] = 14, 10
```

## Location and Scale

Comparing different values of $\mu$ and $\sigma$ 

```python
x = np.linspace(-10.0, 10.0, 1000)

plt.fill(x, ss.norm.pdf(x, 0.0, 1.0), label="$\mu = 0, \sigma = 1.0$", c="b", alpha=0.6, lw=3.0)
plt.fill(x, ss.norm.pdf(x, 2.0, 1.0), label="$\mu = 2.0, \sigma = 1.0$", c="r", alpha=0.6, lw=3.0)
plt.fill(x, ss.norm.pdf(x, 0.0, 2.0), label="$\mu = 0, \sigma = 2.0$", c="g", alpha=0.6, lw=3.0)
plt.fill(x, ss.norm.pdf(x, -1.0, 0.6), label="$\mu = -1.0, \sigma = 0.6$", c="y", alpha=0.6, lw=3.0)


plt.legend()
plt.title("Normal Distribution for different $\mu$ and $\sigma$")
```

## Sampling Distribution

```python
np.set_printoptions(formatter={'float': lambda x: "{0:6.3f}".format(x)})

sampsize = 10
no_samp = 1000
samp = np.random.standard_normal((no_samp,sampsize))
print(samp)
```

```python
mean = samp.sum(axis=1)/sampsize
print(mean)
```
